```python
# This is an example of how Python's built-in csv module works as an alternative to the use of pandas (However, keep in mind, in real-world
# data science and AI workflows, pandas is the standard approach).

import csv

names = []

with open("../data/names.csv", newline="") as file:
    reader = csv.DictReader(file)

    for row in reader:
        names.append(row["name"])

print(names)
```
